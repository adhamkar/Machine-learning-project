```python
import numpy as np
import pandas as pd
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("parkinsons.csv")
```

```python
data
```

```python
data.drop(["name"],axis="columns",inplace=True)
data
```

```python
x=data.drop("status",axis=1)
y=data["status"]
```

```python
x
```

```python
y
```

```python
data.isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=100000)
decisionTreeModel = DecisionTreeClassifier(max_depth=50)
svm_model = LinearSVC(max_iter=10000000)
rnd_forest = RandomForestClassifier(n_estimators=10)
model.fit(X_train, y_train)
decisionTreeModel.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
rnd_forest.fit(X_train, y_train)
```

```python
score = model.score(X_test, y_test)
score2 = decisionTreeModel.score(X_test, y_test)
score3 = svm_model.score(X_test, y_test)
score4 = rnd_forest.score(X_test, y_test)
score, score2, score3, score4
```

```python
scores = cross_val_score(model, x, y, cv=5)
scores2 = cross_val_score(decisionTreeModel, x, y, cv=5)
scores3 = cross_val_score(svm_model, x, y, cv=5)
scores4 = cross_val_score(rnd_forest, x, y, cv=5)
print("Score de validation croisée pour la regression logistique:", np.mean(scores))
print("Score de validation croisée pour l'arbre de decision:", np.mean(scores2))
print("Score de validation croisée pour l'SVM:", np.mean(scores3))
print("Score de validation croisée pour l'arbre aleatoire:", np.mean(scores4))
```
